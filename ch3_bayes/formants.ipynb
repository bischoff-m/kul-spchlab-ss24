# The Formant Triangle



```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tick
plt.rcParams["figure.figsize"] = (10,5)
import librosa
```

### The F1-F2 plane and the formant triangle
Since the days of the Peterson-Barney experiments it has been popular practice to show that steady state vowels 
have quite distinctive representations when represented by their first and second formant .

Plain data exploration is done in notebook [DataExploration.ipynb]()

```python
iy = [345,2400]
ah = [890,1200]
uw = [390,970]
```

```python
f,ax=plt.subplots(figsize=(8,8))
ax.scatter(iy[0],iy[1],100,c='b',marker='*')
ax.scatter(ah[0],ah[1],100,c='g',marker='D')
ax.scatter(uw[0],uw[1],100,c='r',marker='o')


ax.set_xlim([0,1400])
ax.set_ylim([500,3500])
ax.set_yscale('log')
ax.set_yticks([])
ax.yaxis.set_ticklabels([])
ax.set_yticks([600,1000,1500,2000,3000])
y_fmt = tick.FormatStrFormatter('%.0f')
ax.yaxis.set_major_formatter(y_fmt)
ax.set_xlabel('F1')
ax.set_ylabel('F2');
```

```python
ax.text(iy[0]+25,iy[1],'iy',c='b',size=20,va='bottom',ha='left')
ax.text(ah[0]+25,ah[1],'aa',c='g',size=20,va='bottom',ha='left')
ax.text(uw[0]+25,uw[1],'uw',c='r',size=20,va='bottom',ha='left')
display(f)
```
