```python
# do all the imports
%matplotlib inline
import sys, os
import numpy as np
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

#### Code for making the dtw_constraints figures

```python
mat1 = np.zeros((3,3))
fig,axes = plt.subplots(1,2, figsize=(12,4) )
args= dict(cmap='Blues',cbar=False,linecolor='blue',lw=2,square=True,xticklabels=False,yticklabels=False)
ptr_args={'linewidth':1,'head_width':.1,'head_length':.1,
               'length_includes_head':True}
ax = axes[0]
sns.heatmap(mat1,ax=ax,**args)
dxy=.9
ax.arrow(1.5,1.5,dxy,dxy,**ptr_args);
ax.arrow(1.5,2.5,dxy,0,**ptr_args);
ax.arrow(2.5,1.5,0,dxy,**ptr_args);
ax.text(2.5,2.5,"(i,j)",ha="center",va="center")
ax.set_title("DTW Basic Constraints\n Infinite Warping Possible")
###
ax = axes[1]
sns.heatmap(mat1,ax=ax,**args)
dxy=.9
ax.arrow(1.5,1.5,dxy,dxy,**ptr_args);
#ax.arrow(1.5,2.5,dxy,0,**ptr_args);
#ax.arrow(2.5,1.5,0,dxy,**ptr_args);
ax.text(2.5,2.5,"(i,j)",ha="center",va="center")
#
ax.plot([1.5,2.5],[.5,1.5],color='k')
ax.arrow(2.5,1.5,0,dxy,color='k',**ptr_args);
ax.text(1.5,.4,"(i-1,j-2)",ha="center",va="center",color='k')

ax.plot([.5,1.5],[1.5,2.5],color='k')
ax.arrow(1.5,2.5,dxy,0,color='k',**ptr_args);

ax.text(.5,1.4,"(i-2,j-1)",ha="center",va="center",color='k')
ax.set_title("DTW Symmetric Itakura Constraints\n Maximum Warping = x2")
plt.close()
fig
```
